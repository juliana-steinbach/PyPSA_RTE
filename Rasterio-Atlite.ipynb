```python
import rasterio
from rasterio.windows import from_bounds
import pyproj

# Define bounding box in WGS84 (lat/lon) coordinates for France
# These will be converted to ETRS89LAEA CRS
left, bottom, right, top = -5.0, 41.0, 9.5, 51.5  # Approximate coordinates for France (lat/lon)

with rasterio.open(raster_path) as clc:
    # Check CRS
    print(f"Raster CRS: {clc.crs}")

    # Convert the bounding box from lat/lon (WGS84) to ETRS89LAEA using pyproj
    transformer = pyproj.Transformer.from_crs("EPSG:4326", clc.crs.to_string(), always_xy=True)
    left, bottom = transformer.transform(left, bottom)
    right, top = transformer.transform(right, top)

    print(f"Transformed bounding box: left={left}, bottom={bottom}, right={right}, top={top}")

    # Calculate the window from the transformed bounding box
    window = from_bounds(left, bottom, right, top, clc.transform)
    print(f"Window: {window}")

    # Read the data within the window
    band = clc.read(1, window=window)
    print(f"Band shape: {band.shape}")

    # Save the cropped raster to a new file
    output_path = "france_cropped_final.tif"

    # Copy the original CRS and metadata
    output_meta = clc.meta.copy()
    output_meta.update({
        'crs': clc.crs,  # Make sure CRS is copied correctly
        'transform': clc.window_transform(window),  # Transform for the cropped window
        'width': window.width,  # Set width based on the window size
        'height': window.height  # Set height based on the window size
    })

    # Save the cropped raster
    with rasterio.open(output_path, 'w', **output_meta) as dest:
        dest.write(band, 1)

    print(f"Cropped raster saved to {output_path}")
```

```python
A = cutout.availabilitymatrix(shape, excluder)
```

```python
fig, ax = plt.subplots()
A.sel(name="FR").plot(cmap="Greens")
shape.to_crs(4326).plot(ax=ax, edgecolor="k", color="none")
cutout.grid.plot(ax=ax, color="none", edgecolor="grey", ls=":")
```

```python
import xarray as xr

cap_per_sqkm = 2  # 2 MW/km^2
area = cutout.grid.set_index(["y", "x"]).to_crs(3035).area / 1e6  # in km^2
area = xr.DataArray(area, dims=("spatial"))

capacity_matrix = A.stack(spatial=["y", "x"]) * area * cap_per_sqkm
```

```python
wind = cutout.wind(matrix=capacity_matrix, turbine="Vestas_V90_3MW", index=shape.index)
```

```python
wind.plot()
```

```python
pv = cutout.pv(
    panel="CSi",
    orientation={"slope": 30.0, "azimuth": 180.0},
    matrix=capacity_matrix,
    index=shape.index,
)
```
